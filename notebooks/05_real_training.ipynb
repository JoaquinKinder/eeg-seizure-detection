```python
# 1. Primero las magias de Jupyter para recargar módulos
%load_ext autoreload
%autoreload 2

import sys
import os

# 2. Le decimos a Python que mire una carpeta más arriba (¡Esto tiene que ir ANTES de importar src!)
sys.path.append(os.path.abspath('../'))

# 3. Recién ahora importamos nuestras funciones y librerías externas
import numpy as np
from src.data.process_batch import process_patient_batch

# 4. Arranca el procesamiento masivo
print("=== FASE 1: PROCESANDO DATOS CRUDOS ===")
DATA_DIR = "../data/raw"
os.makedirs("../data/processed/", exist_ok=True)

pacientes = ["chb01", "chb02", "chb03", "chb04", "chb05"]

for p_id in pacientes:
    print(f"Masticando paciente {p_id}...")
    X_p, y_p = process_patient_batch(DATA_DIR, p_id)
    
    if X_p is not None:
        np.save(f"../data/processed/X_{p_id}_completo.npy", X_p)
        np.save(f"../data/processed/y_{p_id}_completo.npy", y_p)
        print(f"--> [ÉXITO] {p_id} procesado y guardado en disco.\n")
```

```python
from src.models.balance import get_balance_pipeline
from src.models.classifiers import get_random_forest_classifier

print("=== FASE 2: ENTRENAMIENTO SIN RESTRICCIONES ===")

# 1. Cargamos el bloque de entrenamiento (Pacientes 2 al 5)
X_list, y_list = [], []
for p_id in ["chb02", "chb03", "chb04", "chb05"]:
    X_list.append(np.load(f"../data/processed/X_{p_id}_completo.npy"))
    y_list.append(np.load(f"../data/processed/y_{p_id}_completo.npy"))

X_train_ampliado = np.vstack(X_list)
y_train_ampliado = np.concatenate(y_list)
print(f"-> Volumen total de Train: {X_train_ampliado.shape} ventanas.")

# 2. Balanceo directo 
print("\n-> Aplicando SMOTE...")
balance_pipe = get_balance_pipeline(smote_strategy=0.1, undersample_strategy=0.5, random_state=42)
X_train_bal, y_train_bal = balance_pipe.fit_resample(X_train_ampliado, y_train_ampliado)

# 3. Entrenamiento con 200 árboles
print("\n-> Entrenando Random Forest...")
clf_pesado = get_random_forest_classifier(n_estimators=200, max_depth=20, random_state=42)
clf_pesado.fit(X_train_bal, y_train_bal)
print("   ¡Modelo pesado entrenado con éxito!")

# 4. Predicción sobre nuestro paciente de Test
X_test = np.load("../data/processed/X_chb01_completo.npy") # Cargamos el paciente 1 entero
y_test = np.load("../data/processed/y_chb01_completo.npy")
probabilidades_test = clf_pesado.predict_proba(X_test)[:, 1]
```
